## Mitsuba Emitter Experiment 1

```python
import mitsuba              as mi
import matplotlib.pyplot    as plt
mi.variants()
```

```python
mi.set_variant("llvm_ad_rgb") # mi.set_variant("scalar_rgb") # mi.set_variant("cuda_ad_rgb")
```

## Scenes

We are using two scenes, the _spotbox_ and the _emitbox_ to study different aspects of raytracing.


```python
scene   = mi.load_file("../scenes/emitbox.xml")
img1    = mi.render(scene, spp=128)

plt.axis("off") # plt.figure(figsize = (10,10)) # plt.rcParams["figure.figsize"] = (20,10)
plt.imshow(img1); # approximate sRGB tonemapping  ** (1.0 / 1.7
```

```python
scene   = mi.load_file("../scenes/spotbox.xml")
img1    = mi.render(scene, spp=1024)
plt.axis("off")
plt.imshow(img1 ** (1.0 / 1.7))
```

```python
scene   = mi.load_file("../scenes/diskbox.xml")
img1    = mi.render(scene, spp=128)
plt.axis("off")
plt.imshow(img1 ** (1.0 / 1.7))
```

## Updating the scene

In this example, the color of the emitter is updated.

```python
scene   = mi.load_file("../scenes/diskbox.xml")
params = mi.traverse(scene)
# print(params)
#print(params["SpotLight.intensity.value"])

params['SpotLight.intensity.value'] = [1., 10., 10.]
params.update();
img1    = mi.render(scene, spp=128)
plt.axis("off")
plt.imshow(img1 ** (1.0 / 1.7))
```
