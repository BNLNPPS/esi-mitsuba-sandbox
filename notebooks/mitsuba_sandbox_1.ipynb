```python
import mitsuba as mi
```

```python
mi.variants()
```

```python
# mi.set_variant("llvm_ad_rgb")
# mi.set_variant("scalar_rgb")
mi.set_variant("cuda_ad_rgb")
```

```python
# mi.set_variant('scalar_rgb')

img1 = mi.render(mi.load_dict(mi.cornell_box()))
```

```python
import matplotlib.pyplot as plt

plt.axis("off")
plt.imshow(img1 ** (1.0 / 2.2)); # approximate sRGB tonemapping
```
