## Mitsuba Disk Experiment 1

This notebook is used to test out the details of the setup, which goes as follows:
* The scene is delimited by a cubic room
* Ther is a dim light source on top to ascertain the presence of some optional elements we may add to the scene for experimentation
* The sensor is located on the front face of the room i.e. where the sernsor (observer) is
* The light source is a "spot", which is an emitter producing cone-shaped raidation forward
* The disk in the back part of the scene is tilted
* The purpose is to validate the scattered light and reflections, e.g. by implementing a mirror surface on the bottom.


```python
import mitsuba              as mi
import matplotlib.pyplot    as plt
mi.variants() # mi.set_variant("llvm_ad_rgb") # mi.set_variant("scalar_rgb") # mi.set_variant("cuda_ad_rgb")
# mi.set_variant("scalar_rgb")
mi.set_variant("llvm_ad_rgb")
```

```python
scene   = mi.load_file("../scenes/diskbox.xml")
img1    = mi.render(scene, spp=256)


# plt.figure(figsize=(10,10))
plt.imshow(img1 ** (1.0 / 1.7))
plt.axis("off")
```

```python
scene   = mi.load_file("../scenes/sidebox.xml")
img1    = mi.render(scene, spp=256)


# plt.figure(figsize=(10,10))
plt.imshow(img1 ** (1.0 / 1.7))
plt.axis("off")
```

## Updating the scene (Optional)
